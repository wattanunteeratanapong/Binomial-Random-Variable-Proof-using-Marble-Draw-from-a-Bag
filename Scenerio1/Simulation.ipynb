```python
# Scenerio 1
# There are 2 marbles in a bag: 1 red, 1 green.

# Success : If the color of the marble is red.
# Failure : If the color of the marble is green.
```

```python
import random
import matplotlib.pyplot as plt
```

```python
trial = 10  
bag = ['red', 'green', 'green']  
record = []  
```

```python
def simulation(time):
    for _ in range(time):
        success = 0
        for _ in range(trial):
            random_marble = random.choice(bag)
            if random_marble == 'red':
                success += 1
        record.append(success)
```

```python
simulation(10000000)
```

```python
plt.hist(record, bins=range(0, trial+2), align='left', edgecolor='black')
plt.xlabel('Number of Successes (Red draws)')
plt.ylabel('Frequency')
plt.title(f'Distribution of Successes in 10000000 Simulations')
plt.show()
```

```python
plt.hist(record, bins=range(0, trial+2), align='left', edgecolor='black', density=True)
plt.xlabel('Number of Successes (Red draws)')
plt.ylabel('Probability')
plt.title('Normalized Distribution of Successes in 10000000 Simulations')
plt.show()
```

```python
total_simulations = len(record)
outcomes = list(range(trial + 1))
probabilities = [record.count(outcome) / total_simulations for outcome in outcomes]

print("Probabilities:")
for outcome, prob in zip(outcomes, probabilities):
    print(f'P({outcome} {trial}) = {prob:.8f}')
```
